# Simulation of Low Earth OrbitObjects

### Importing the adequate modules

For this tutorial, we will import the `Database_Simulation` and `datetime` modules

```python
import datetime
import Database_Simulation as ds
```

### Preprocessing a 3LE .txt file

```python
filename = '3le_all.txt' # See '3LE load informations.txt' to see how to download 3le objects in LEO
nb_satellites = 20 # The number of unique objects to keep at the end - Optional 

satellites = ds.process3LE(filename)[:nb_satellites] # Retrieves unique objects from the 3LE. Remove "nb_satellites" to get all.
```

### Initialize the dataset

```python
duration = 1  # Number of hours in the simulation
sec_step = 5  # Number of seconds per time step
start = datetime.datetime.now()  # Start of simulation

data = ds.initializeDatabase(satellites, start, duration=1, sec_step=5)
```

```python
data
```

### Save the dataset

Accepted formats : `csv`, `json`, `excel`. The format is specified as the extension of the saved file, `json` in this tutorial.

```python
path = 'Data/satellites_example.json' # Path to save the database. Can be an absolute path or one starting from the current folder.

ds.saveDataframe(data, path)
```

Reads data from a specified file.

```python
data = ds.readsDataframe(path)
data
```

### Insert a specified object in the database

Takes a .txt file of a 3LE as parameter, only one object at a time.

```python
file = 'single_3le.txt'

new_data = ds.insert(file, data)
new_data
```

### Search for a specific satellite

```python
row = ds.searchByName("EXPLORER 1", data)
row
```

### Delete a list of objects by name

```python
list_to_delete = ['EXPLORER 1', 'AAAAA']

new_data = ds.delete(list_to_delete, data)
new_data
```

### Update the database

Updates every object in the database for a given number of seconds, starting for each object at the last computation of nearest neighbour time.

```python
seconds_to_add = 200

ds.updateDatabase(data, seconds_to_add)
```
